***Question 7*** 

Write a program using Nested if statement for ATM withdrawal.                              
Conditions:                                                
* First check PIN 
* Then check balance 
* If both are correct → Print "Transaction Successful" 
* Otherwise print proper message.

```python
correct_pin = "1234"
balance = 5000

correct_pin = int(input("enter you pin"))
amount = int(input("enter you withdrawal amount"))
if correct_pin == 1234:
    if balance >= amount:
        print("Transaction Successful")
    else:
        print("Transaction Fail")
else:
    print("Invalid pin")
```
