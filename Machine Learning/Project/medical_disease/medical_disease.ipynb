```python
disease_count = df['Disease'].value_counts()
label = ['Healthy', 'Pre-Diabetes', 'Hypertension', 'Heart Disease', 'Diabetes']
plt.pie(disease_count,labels = label , autopct = "%1.1f%%")
plt.show()
```

```python
sns.countplot(x='Disease', data=df , color= 'Purple')
plt.title("Distribution of Disease Cases")
plt.xlabel("Disease Status")
plt.ylabel("Number of Patients")
plt.show()
```

```python
sns.countplot(data = df , x = df['Disease'],hue = df['Gender'],palette='Set1')
plt.title("Gender wise Disease Distribution")
plt.xlabel("Disease Type", fontsize=12)
plt.ylabel("Count of Patients", fontsize=12)
```

```python
sns.countplot(x='FamilyHistory', hue='Disease', data=df, palette='Set1')
plt.title("Impact of Family History on Disease Occurrence")
plt.xlabel("Family History")
plt.ylabel("Number of Patients")
plt.show()
```

```python
df['Gender'].unique()
```

```python

Gender_dict = {'Male':1 , 'Female':2 }
df['Gender'] = df['Gender'].map(Gender_dict)
```

```python
df['Smoking'].unique()
```

```python
Smoking_dict = {'Yes' :1 , 'No':2}
df['Smoking'] = df["Smoking"].map(Smoking_dict)
```

```python

df["Alcohol"].unique() 
```

```python
Alcohol_dict = {'Yes':1,'No': 2}
df['Alcohol'] = df['Alcohol'].map(Alcohol_dict)

```

```python

df["PhysicalActivity"].unique()
```

```python
PhysicalActivity_dict = {'High':1 ,'Medium':2,'Low':3}
df['PhysicalActivity'] = df['PhysicalActivity'].map(PhysicalActivity_dict)
     
```

```python
df["FamilyHistory"].unique() 
```

```python
Family_History_dict = {'Yes' : 1 , 'No' : 2 }
df['FamilyHistory'] = df['FamilyHistory'].map(Family_History_dict)
```

```python
disease_dict = {'Healthy':1, 'Pre-Diabetes':2, 'Hypertension':3, 'Heart Disease':4, 'Diabetes':5}
df['Disease'] = df['Disease'].map(disease_dict)
```
