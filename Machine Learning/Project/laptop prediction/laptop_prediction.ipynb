```python
sns.histplot(df['price'],kde=True)
plt.xlabel('Price+')
plt.ylabel('Number of laptops')
plt.title('Number of Laptops')
```

```python
sns.barplot(x=df['brand'],y=df['price'],errorbar=None)  
plt.title('Brand vs Price')
plt.xticks(rotation=90)

plt.show()
```

```python
sns.barplot(x=df['CPU'],y=df['price'],errorbar=None)
plt.xlabel("Number of Cores")
plt.ylabel("Average Price")
plt.title("Price VS CPU core count")
plt.xticks(rotation=90)
plt.show()


```

```python
sns.barplot(x=df['Ram'],y=df['price'],data=df,errorbar=None)
plt.xlabel("RAM")
plt.ylabel("Average Price")
plt.title("RAM VS Price")
plt.show()
```

```python
sns.barplot(x=df['Ram_type'],y=df['price'],errorbar=None)
plt.xlabel("Type of Ram ")
plt.ylabel("Price")
plt.title("Price disribution among Ram types ")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(x=df['ROM'],y=df['price'],errorbar=None)
plt.xlabel('ROM')
plt.ylabel('Price')
plt.title('ROM VS Price')
plt.show()
```

```python
sns.boxplot(x='OS',y='price',data=df)
plt.xlabel("Operating System")
plt.ylabel("Price")
plt.xticks(rotation=90)
plt.show()
```
