1. Create a class Employee with attributes emp_name, emp_id, and salary.

```python
class Employee:
    def __init__(self, emp_name, emp_id, salary):
        self.emp_name = emp_name
        self.emp_id = emp_id
        self.salary = salary

    def display(self):
        print("Employee Name:", self.emp_name)
        print("Employee ID:", self.emp_id)
        print("Salary:", self.salary)
        print()

# Creating 3 objects
e1 = Employee("Akanksha", 101, 50000)
e2 = Employee("Rahul", 102, 45000)
e3 = Employee("Priya", 103, 55000)

# Display details
e1.display()
e2.display()
e3.display()
```
