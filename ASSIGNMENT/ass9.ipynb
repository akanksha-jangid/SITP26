1. Create a NumPy array of 50 random numbers and find the 2nd largest and 3rd smallest element

```python
import numpy as np
arr=np.random.randint(1,100,50)
sorted_arr=np.sort(arr)
second_largest=sorted_arr[-2]
third_smallest=sorted_arr[2]

print("Array:",arr)
print("2nd Largest:", second_largest)
print("3rd Smallest:",third_smallest)
```

2. Given a DataFrame, identify all columns having more than 20% missing 
values and remove them. 

```python
import numpy as np
import pandas as pd
df=pd.DataFrame({
    'A':[1,np.nan,3,np.nan,5],
    'B':[10,20,30,40,50],
    'C':[np.nan,np.nan,np.nan,40,50],
    'D':[1,2,3,4,np.nan]
})

print("Original Dataframe:")
print(df)

missing_percentage=(df.isnull().sum()/len(df))*100
columns_to_remove=missing_percentage[missing_percentage>20].index
df=df.drop(columns=columns_to_remove)
print('\nColumns Removed:',list(columns_to_remove))
print("\nDataFrame After Removing Columns:")
print(df)
```

3. Create a NumPy matrix and find its transpose, determinant, rank, and 
inverse. 

```python
import numpy as np
matrix = np.array([[1,2,3],[0,1,4],[5,6,0]])
print("Transpose:\n",matrix.T)
print("Determinant:",np.linalg.det(matrix))
print("Rank:",np.linalg.matrix_rank(matrix))
print("Inverse:\n",np.linalg.inv(matrix))
```

4. Using Pandas, find the top 5 categories with the highest average value 
of a numerical column.  

```python
import pandas as pd

df = pd.DataFrame({
    'Category':['A','B','A','C','B','D','E','A','C','D'],
    'Value':[10,20,15,30,25,40,35,18,28,50]
})
result = df.groupby('Category')['Value'].mean().sort_values(ascending=False).head(5)
print(result)
```

5. Create a Seaborn heatmap and identify the three most highly 
correlated feature pairs.  

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame(np.random.rand(100,5),
                  columns=['A','B','C','D','E'])

corr = df.corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
corr_pairs=corr.unstack()
corr_pairs=corr_pairs[corr_pairs !=1]
top3 = corr_pairs.abs().sort_values(ascending=False).drop_duplicates().head(3)
print("Top 3 Correlated Pairs:")
print(top3)
```

6. Generate a dataset with NumPy and visualize its distribution using 
Histogram, KDE Plot, and Box Plot.  

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data=np.random.normal(50,10,1000)

plt.hist(data,bins=20)
plt.title("Histogram")
plt.show()
sns.kdeplot(data, fill=True)
plt.title("KDE Plot")
plt.show()

sns.boxplot(x=data)
plt.title("Box Plot")
plt.show()
```

7. Using Pandas GroupBy, find the highest, lowest, and average sales for 
each category. 

```python
import pandas as pd

df = pd.DataFrame({
    'Category':['A','A','B','B','C','C'],
    'Sales':[100,150,200,180,250,300]
})

result=df.groupby('Category')['Sales'].agg(
    Highest='max',
    Lowest='min',
    Average='mean'

)

print(result)
```

8. Detect outliers in a numerical column using the IQR method and 
visualize them using a Box Plot.  

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Value':[10,12,15,18,20,22,25,100]
})

Q1= data['Value'].quantile(0.25)
Q3= data['Value'].quantile(0.75)

IQR = Q3 -Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers = data[(data['Value'] < lower) | (data['Value'] > upper)]

print("Outliers:")
print(outliers)

sns.boxplot(x=data['Value'])
plt.show()
```

9. Create a Matplotlib figure containing 4 subplots:  
o Line Plot  
o Bar Plot  
o Scatter Plot  
o Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,11)
y = x**2

plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("Line Plot")

plt.subplot(2,2,2)
plt.bar(x,y)
plt.title("Bar Plot")

plt.subplot(2,2,3)
plt.scatter(x,y)
plt.title("Scatter Plot")

plt.subplot(2,2,4)
plt.hist(y)
plt.title("Histogram")

plt.tight_layout()
plt.show()
```

10. Create a rolling average of window size 7 on a dataset and compare 
it with the original trend using Matplotlib. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.Series(np.random.randint(50,100,30))

rolling_avg=data.rolling(window=7).mean()
plt.plot(data,label='Original')
plt.plot(rolling_avg,label='Rolling Average')
plt.legend()
plt.show()
```

12. Build a mini EDA report that includes:  
● Missing Values Analysis  
● Correlation Analysis  
● Outlier Detection  
● Feature Distribution Visualization 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Hp\Desktop\data.csv")

print("Missing Values:")
print(df.isnull().sum())

corr = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()

num_cols = df.select_dtypes(include=np.number).columns

for col in num_cols:
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot - {col}")
    plt.show()

for col in num_cols:
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution - {col}")
    plt.show()
```
