**Inheritance in Python**

Inheritance is an OOPs (Object-Oriented Programming) concept in Python where one class can use the properties and methods of another class.                                  

* The class whose properties are inherited is called Parent Class / Base Class
* The class that inherits is called Child Class / Derived Class                                    

 Inheritance helps in:

* Code reusability
* Reducing duplicate code
* Easy maintenance


**Types of Inheritance in Python**


| Type                     | Meaning                                    |
| ------------------------ | ------------------------------------------ |
| Single Inheritance       | One parent → One child                     |
| Multiple Inheritance     | Multiple parents → One child               |
| Multilevel Inheritance   | Grandparent → Parent → Child               |
| Hierarchical Inheritance | One parent → Multiple children             |
| Hybrid Inheritance       | Combination of different inheritance types |

2. Create a class PrimeNumberChecker with a function to check whether a 
number is prime or not. 

```python
class PrimeNumberChecker:
    def check_prime(self):
        num = int(input("Enter a number: "))

        if num > 1:
            for i in range(2, num):
                if num % i == 0:
                    print(num, "is not a Prime Number")
                    break
            else:
                print(num, "is a Prime Number")
        else:
            print(num, "is not a Prime Number")


obj = PrimeNumberChecker()


obj.check_prime()
```

* ***QUESTION 7***Create a class BankAccount with functions: 
* deposit() 
* withdraw() 
* check_balance() 


```python
class BankAccount:
    def __init__(self):
        self.balance = 0
    def deposit(self):
        amount = float(input("Enter amount to deposit: "))
        self.balance += amount
        print("Amount Deposited :", amount)
    def withdraw(self):
        amount = float(input("Enter amount to withdraw: "))

        if amount <= self.balance:
            self.balance -= amount
            print("Amount Withdrawn :", amount)
        else:
            print("Insufficient Balance")
    def check_balance(self):
        print("Current Balance :", self.balance)
obj = BankAccount()
obj.deposit()
obj.withdraw()
obj.check_balance()
```

**Constuctor with Inheritance**

**super().__init__(name, age)**                                               
Ye parent class constructor ko call karta hai.

Internally:
***Person.__init__(self, name, age)***

```python
# Parent Class
class Person:

    def __init__(self, name, age):
        self.name = name
        self.age = age


# Child Class
class Teacher(Person):

    def __init__(self, name, age, subject, salary):

        # Parent constructor call
        super().__init__(name, age)

        # Child class variables
        self.subject = subject
        self.salary = salary

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Subject:", self.subject)
        print("Salary:", self.salary)


# Object Creation
obj = Teacher("Akanksha", 20, "Python", 50000)

obj.display()
```

***Question5***Create a class Calculator with functions:       
○ add()                              
○ subtract()                        
○ multiply()                     
○ divide()                

```python
class Calculator:
    
    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

    def multiply(self, a, b):
        return a * b

    def divide(self, a, b):
        if b == 0:
            return "Division by zero is not allowed"
        return a / b

calc = Calculator()
print("Addition:", calc.add(10, 5))
print("Subtraction:", calc.subtract(10, 5))
print("Multiplication:", calc.multiply(10, 5))
print("Division:", calc.divide(10, 5))
```

Create a parent class Person and child class Employee using single 
inheritance. Display employee details. 

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
class Employee(Person):
    
    def __init__(self, name, age, emp_id, salary):
        # Calling parent class constructor
        super().__init__(name, age)
        
        self.emp_id = emp_id
        self.salary = salary

    def display(self):
        print("Employee Details")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Employee ID:", self.emp_id)
        print("Salary:", self.salary)
e1 = Employee("Akanksha", 20, 101, 50000)
e1.display()
```
