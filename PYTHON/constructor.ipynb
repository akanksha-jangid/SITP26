```python
class ATM:
    def __init__(self,bal):
        self.bal = bal
        print("ATM Started")
        print("Current Balance:", self.bal)
    
    def check_balance(self):
        print("Your Balance is:", self.bal)

    def withdrwa(self,amount):
        if amount <= self.bal:
            self.bal -= amount
            print(amount, "withdrawn successfully")
            print("Remaining Balance:", self.bal)
        else:
            print("Insufficient Balance")



a = ATM(10000)
```

```python
class Employee:
    def __init__(self, emp_name, emp_id, salary):
        self.emp_name = emp_name
        self.emp_id = emp_id
        self.salary = salary

    def display(self):
        print("Employee Name:", self.emp_name)
        print("Employee ID:", self.emp_id)
        print("Salary:", self.salary)
        print()
e1 = Employee("Akanksha", 101, 50000)
e2 = Employee("Rahul", 102, 45000)
e3 = Employee("Priya", 103, 55000)

e1.display()
e2.display()
e3.display()
```

2. Create a class Calculator with methods for addition, subtraction, multiplication, and division.

```python
class Calculator:
    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

    def multiply(self, a, b):
        return a * b

    def divide(self, a, b):
        if b != 0:
            return a / b
        return "Division by zero not allowed"

c = Calculator()

print("Addition:", c.add(10, 5))
print("Subtraction:", c.subtract(10, 5))
print("Multiplication:", c.multiply(10, 5))
print("Division:", c.divide(10, 5))
```

3. Create a class Book with a constructor to initialize title, author, and price, then print details.

```python
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price

    def display(self):
        print("Title:", self.title)
        print("Author:", self.author)
        print("Price:", self.price)

b1 = Book("Python Basics", "ABC", 499)
b1.display()
```
