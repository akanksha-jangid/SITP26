**Class in Python**

A class is a blueprint or template used to create objects.                          

It contains:                                                         

Variables (attributes/data)                                         
Functions (methods/behaviors)                                         

Real-life example:                                                           

Class = Car design                                                      
Object = Actual car                                 

**Pillars of OOP (Object-Oriented Programming)**

***There are 4 main pillars of OOP:***

* Encapsulation
* Inheritance
* Polymorphism
* Abstraction

1. **Encapsulation**

Encapsulation means wrapping data and methods into one unit (class).                                                         

It also helps in data hiding.                                      

2. **Inheritance**

Inheritance means one class can use properties of another class.                                     

Parent class → Base class                                             
Child class → Derived class                                             

3. **Polymorphism**

Polymorphism means one thing having many forms.                                            

Same function behaves differently.                                              

4. **Abstraction**

Abstraction means showing important details and hiding internal implementation.
