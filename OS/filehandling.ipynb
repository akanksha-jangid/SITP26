File Handling in Python ka use files ko create, read, write, update aur delete karne ke liye hota hai.
Jaise .txt, .csv, etc. files ke data ko handle karna.

| Mode   | Meaning                                  |
| ------ | ---------------------------------------- |
| `"r"`  | Read file                                |
| `"w"`  | Write file (old data delete ho jata hai) |
| `"a"`  | Append data at end                       |
| `"x"`  | New file create                          |
| `"r+"` | Read + Write                             |


file = open("filename.txt", "mode")
* filename.txt" → file ka naam
* mode" → file ko kis purpose se open karna hai

```python
file =open("poit.txt",'r')
```

```python
file.read()
```

```python
#file.readline()
#file.readline()
file.close()
```

```python
with open('poit.txt') as file:
    print(file.read())
```

```python
file=open('poit.txt','w')
file.write("I am from sikar")
file.close()
```

```python
file=open('poit.txt','r')
file.read()
```

```python
file=open('poit.txt','a')
file.write("File Handling in python means working with files like creating,reading,writing,updating etc.")
file.close()
```
