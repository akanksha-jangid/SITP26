**CONSTRUCTOR**

```python
class ATM:
    def __init__(self,bal):
        self.bal = bal
        print("ATM Started")
        print("Current Balance:", self.bal)
    
    def check_balance(self):
        print("Your Balance is:", self.bal)

    def withdrwa(self,amount):
        if amount <= self.bal:
            self.bal -= amount
            print(amount, "withdrawn successfully")
            print("Remaining Balance:", self.bal)
        else:
            print("Insufficient Balance")



a = ATM(10000)
```

```python

a.check_balance()
```

```python
a.withdrwa(3000)
```

```python
a.check_balance()
```

```python
class BMI:
    def __init__(self,weight,height):
        self.height=height
        self.weight=weight
        self.calculator()
            
    def calculator(self):
        bmi = self.weight/(self.height*self.height)
        print("BMI : ", bmi)

person1 = BMI(4,98)
```
