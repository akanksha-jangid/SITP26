* QUESTION 1.Create a class NumberSeries that  prints numbers from 1 to 200 using a 
loop. 
 

```python
class Numberseries:
    def print_number(self):
        for i in range(1,201):
            print(i)
obj = Numberseries()
obj.print_number()
```

2. Create a class PrimeNumberChecker with a function to check whether a 
number is prime or not. 

```python
class PrimeNumberChecker:
    def check_prime(self):
        num = int(input("Enter a number: "))

        if num > 1:
            for i in range(2, num):
                if num % i == 0:
                    print(num, "is not a Prime Number")
                    break
            else:
                print(num, "is a Prime Number")
        else:
            print(num, "is not a Prime Number")


obj = PrimeNumberChecker()


obj.check_prime()
```

3. Create a class Student with using constructor to store: 
* name 
* roll number 
* course 

```python
class Student:
    def __init__(self, name, roll_number, course):
        self.name = name
        self.roll_number = roll_number
        self.course = course

    def display(self):
        print("Name:", self.name)
        print("Roll Number:", self.roll_number)
        print("Course:", self.course)
student1 = Student("Akanksha", 101, "AI & ML")
student1.display()
```

4. Add a function to display student details.

```python
class Student:

    def get_details(self):
        self.name = input("Enter Student Name: ")
        self.roll_number = int(input("Enter Roll Number: "))
        self.course = input("Enter Course Name: ")
    def show_details(self):
        print("\n--- Student Details ---")
        print("Name :", self.name)
        print("Roll Number :", self.roll_number)
        print("Course :", self.course)
obj = Student()
obj.get_details()
obj.show_details()
```

* **Question8** Create a class FactorialCalculator with a function to calculate the 
factorial of a number using a loop. 

```python
class FactorialCalculator:
    def factorial(self):
        num = int(input("Enter a number: "))
        
        fact = 1

        for i in range(1, num + 1):
            fact = fact * i

        print("Factorial of", num, "is", fact)

obj = FactorialCalculator()
obj.factorial()
```
