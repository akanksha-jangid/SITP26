***for Loop ***
ka use kisi sequence (list, string, tuple, range etc.) ko repeat/iterate karne ke liye hota hai.
   ***Syntax : for variable in sequence:
    statements***

* ***Types of loop***
* for loop
* while loop

***Features***

* variable → ek-ek value store karta hai
* sequence → list, string, range etc.

***uses***
when the itration is known

```python
name = "Akanksha"
for x in name:
    print(x)
```

```python
lst = [1,2,3,4,5]
for y in lst:
    print(y)
```

```python
range(0,20)
for x in range(0,20):
    print(x)
```

```python
for x in range(2,22,2): ###ending 2 is differnce of 2 in printing the range
    print(x)

```

```python
a = {"name":"Akanksha","Address":"Sikar","College":"GWECA"}
for x in a:
    print(x)
```

```python
tpl = (1,3,5,6,7,8,4,3,6,89,5)
for item in tpl:
    print(item)
```

```python
lst = [1,2,3,4]
for x in lst:
    if x == 3:
        break ####break stop the loop on assign value
    print(x)
```

```python
lst = [1,2,3,4 ,5] 
for x in lst:
    if x == 3:
        continue ####continue skips the value assign 
    print(x)
```

```python
lst = ["apple","graps","mango","gavava"]
for y in lst:
    if y == "mango":
        continue
    print(y)
```

***While loop on python***                                     

A while loop in Python repeatedly executes a block of code as long as a condition is True.

while condition:                                
    # code to run

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

```python
count = 1
while count <= 7:
    print(count)
    count += 1
    if count == 6:
        break
```

```python
num = 10

while num >= 1:
    print(num)
    num -= 1
```
