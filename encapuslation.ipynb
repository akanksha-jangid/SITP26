**Encapsulation**                          
Encapsulation in Python                                                                  


Encapsulation ka matlab hota hai data (variables) aur methods (functions) ko ek single unit yani class ke andar bind karna aur data ko direct access se protect karna.                                             


* Simple words mein:                                                                            

“Data hiding + controlled access” = Encapsulation

**Encapsulation Syntax in Python**                                                

Python mein encapsulation ke liye:                                                 

Public variable → normal variable                                    
Protected variable → _variable                                         
Private variable → __variable                                      

```python
class College:
    college_name="GWECA" #############public variable
    college_garden="yes" ########################public variable
    college_account=123546789  #############public variable
    
```

```python
clg=College()
```

```python
clg.college_name
```

```python
clg.college_garden
```

```python
clg.college_account
```

```python
class College:
    college_name="GWECA" #############public variable
    _college_garden="yes" ########################protected variable
    college_account=123546789  #############public variable
```

```python
clg=College()
```

```python
clg.college_name
```

```python
clg._college_garden
```

```python
clg.college_account
```

```python
class College:
    college_name="GWECA" #############public variable
    _college_garden="yes" ########################protected variable
    __college_account=123546789  #############private variable

    def display(self):
        self.__college_account
        print("Account Number will be:",self.__college_account)

clg=College()
```

```python
clg.display()
```

```python
clg._College__college_account
```

```python
clg=College()
```

```python
clg.college_name
```

```python
clg._college_garden
```
